```python
import pandas as pd
health = pd.read_csv("../Datasets/rankmd.csv", delimiter=";")
```

```python
States = health.copy()
States = States[States['FIPS'].astype(str).str.endswith('000')]

import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(3, 2, sharex=False, figsize=(30,37))
sns.set(rc={'figure.figsize':(12,9)})
sns.despine(left=True, bottom=True)
sns.barplot(ax=axes[0,0], data=States.sort_values("premature_death_YPLL Rate (White)",ascending=False), y='State', x='premature_death_YPLL Rate (White)').set(title='White',  xlabel='Potential Life lost rate')
sns.barplot(ax=axes[0,1], data=States.sort_values("premature_death_YPLL Rate (Black)",ascending=False), y='State', x='premature_death_YPLL Rate (Black)').set(title='Black',  xlabel='Potential Life lost rate')
sns.barplot(ax=axes[1,0], data=States.sort_values("premature_death_YPLL Rate (Asian)",ascending=False), y='State', x='premature_death_YPLL Rate (Asian)').set(title='Asian', xlabel='Potential Life lost rate')
sns.barplot(ax=axes[1,1], data=States.sort_values('premature_death_YPLL Rate (Hispanic)',ascending=False), y='State', x='premature_death_YPLL Rate (Hispanic)').set(title='Hispanic', xlabel='Potential Life lost rate')
sns.barplot(ax=axes[2,0], data=States.sort_values('premature_death_YPLL Rate (AIAN)',ascending=False), y='State', x='premature_death_YPLL Rate (AIAN)').set(title='Native American', xlabel='Potential Life lost rate')
```
